```python
import numpy as np
import pandas as pd
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import time

# Check if MPS is available (for Mac), else use CPU
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")

# Pre-trained Models to Compare
models = [
    "gpt2",
    "EleutherAI/gpt-neo-125M",
    "distilgpt2"
]

# Prompt for Text Generation
prompt = "Artificial Intelligence will change the future by"

# Function to Evaluate Models
def evaluate_model(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
    generator = pipeline('text-generation', model=model, tokenizer=tokenizer, device=0 if device.type == "cuda" else -1)

    start_time = time.time()
    output = generator(prompt, max_length=50, num_return_sequences=1)[0]['generated_text']
    end_time = time.time()

    # Metrics
    output_length = len(output.split())
    repetition_rate = len(output.split()) - len(set(output.split()))
    generation_time = end_time - start_time

    return {
        'Model': model_name,
        'Output Length': output_length,
        'Repetition Rate': repetition_rate,
        'Generation Time (s)': generation_time,
        'Generated Text': output[:100] + "..."
    }

# Apply Evaluation
results = [evaluate_model(model) for model in models]
df = pd.DataFrame(results)
print(df)

# Normalize Data for TOPSIS
scaler = MinMaxScaler()
criteria = df[['Output Length', 'Repetition Rate', 'Generation Time (s)']]
normalized_criteria = scaler.fit_transform(criteria)

# Weights (adjust as needed)
weights = np.array([0.4, 0.3, 0.3])  # Prioritize Output Length
ideal_best = [1, 0, 0]               # Higher Output, Lower Repetition & Time
ideal_worst = [0, 1, 1]              # Opposite for worst-case

# TOPSIS Score Calculation
dist_best = np.linalg.norm(normalized_criteria - ideal_best, axis=1)
dist_worst = np.linalg.norm(normalized_criteria - ideal_worst, axis=1)
topsis_score = dist_worst / (dist_best + dist_worst)

# Add Scores to DataFrame
df['TOPSIS Score'] = topsis_score
df['Rank'] = df['TOPSIS Score'].rank(ascending=False).astype(int)
df = df.sort_values(by='TOPSIS Score', ascending=False)
print("\nFinal Rankings:\n", df[['Model', 'TOPSIS Score', 'Rank']])

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='TOPSIS Score', y='Model', data=df, palette='coolwarm')
plt.title('TOPSIS Ranking of Text Generation Models')
plt.xlabel('TOPSIS Score')
plt.ylabel('Model')
plt.show()

# Save Results to CSV for GitHub Upload
df.to_csv("topsis_results.csv", index=False)
```
